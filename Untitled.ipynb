```python
def choose_macbook(price_m4, price_m5, workload, urgency, months_after_launch):
    # 我们给性能和耐用性设定标准数值（你也可以自己改）
    performance_m5 = 100    # M5 性能基准分
    longevity_m5 = 100      # M5 耐用性基准分
    performance_m4 = 90     # M4 性能分
    longevity_m4 = 95       # M4 耐用性分

    # 计算性价比得分（效用值 = 性能 × 耐用性 / 价格）
    m5_score = performance_m5 * longevity_m5 / price_m5
    m4_score = performance_m4 * longevity_m4 / price_m4

    print(f"\nM5 性价比得分：{m5_score:.2f}")
    print(f"M4 性价比得分：{m4_score:.2f}\n")

    # 决策规则（从上到下优先级）
    if urgency == "high":
        return "立即买 M5"
    if months_after_launch < 8:
        return "M5 新机期，推荐买 M5"
    if m4_score > m5_score * 1.15:
        return "M4 性价比更高，推荐买 M4"
    
    return "不着急，建议再观望 WAIT"


# ======================
# 在这里输入数字 👇
# ======================
print("=== MacBook 购买决策工具 ===")
price_m4 = float(input("请输入 M4 价格："))
price_m5 = float(input("请输入 M5 价格："))
workload = input("请输入使用场景（随便输入即可）：")
urgency = input("是否急需？(high / normal)：")
months_after_launch = int(input("M5 发布已过几个月："))

# 调用函数并输出结果
result = choose_macbook(price_m4, price_m5, workload, urgency, months_after_launch)
print("最终决策：", result)
```

```python
"""
MacBook Purchase Decision System
================================

A practical, extensible MacBook recommendation engine
built with:

- Logic tree decision making
- Weighted scoring system
- Workload modeling
- Future-proof analysis
- Budget constraints

Designed for:
- MacBook Air / Pro decisions
- AI workflows
- Programming
- Students
- Creators
- Long-term value analysis

Python 3.11+
"""

from dataclasses import dataclass
from typing import List, Dict


# =========================================================
# DATA MODEL
# =========================================================

@dataclass
class MacBook:
    name: str
    chip: str
    ram: int
    storage: int
    size: int
    fan: bool
    price: int

    cpu_score: int
    gpu_score: int
    portability_score: int
    battery_score: int

    ai_score: int
    multitask_score: int
    thermal_score: int


# =========================================================
# MACBOOK DATABASE
# =========================================================

MACBOOKS = [

    MacBook(
        name="MacBook Air M5 13",
        chip="M5",
        ram=16,
        storage=512,
        size=13,
        fan=False,
        price=6199,

        cpu_score=92,
        gpu_score=88,
        portability_score=98,
        battery_score=96,

        ai_score=78,
        multitask_score=72,
        thermal_score=65,
    ),

    MacBook(
        name="MacBook Air M4 15",
        chip="M4",
        ram=24,
        storage=512,
        size=15,
        fan=False,
        price=8099,

        cpu_score=84,
        gpu_score=80,
        portability_score=78,
        battery_score=94,

        ai_score=92,
        multitask_score=95,
        thermal_score=72,
    ),

    MacBook(
        name="MacBook Air M5 15",
        chip="M5",
        ram=24,
        storage=512,
        size=15,
        fan=False,
        price=9499,

        cpu_score=96,
        gpu_score=92,
        portability_score=76,
        battery_score=95,

        ai_score=98,
        multitask_score=98,
        thermal_score=74,
    ),

    MacBook(
        name="MacBook Pro 14 M5",
        chip="M5",
        ram=16,
        storage=512,
        size=14,
        fan=True,
        price=10999,

        cpu_score=98,
        gpu_score=96,
        portability_score=82,
        battery_score=92,

        ai_score=82,
        multitask_score=78,
        thermal_score=98,
    )
]


# =========================================================
# USER PROFILE
# =========================================================

@dataclass
class UserProfile:
    budget: int

    coding: int
    ai: int
    video_editing: int
    multitasking: int
    portability: int
    gaming: int

    lifespan_years: int


# =========================================================
# LOGIC TREE
# =========================================================

class LogicEngine:

    @staticmethod
    def determine_priorities(user: UserProfile) -> Dict[str, float]:

        weights = {
            "cpu": 0.15,
            "gpu": 0.10,
            "ram": 0.20,
            "portability": 0.15,
            "battery": 0.10,
            "ai": 0.10,
            "multitask": 0.10,
            "thermal": 0.10
        }

        # ==========================================
        # AI workloads
        # ==========================================

        if user.ai >= 8:
            weights["ram"] += 0.15
            weights["ai"] += 0.10
            weights["thermal"] += 0.05

        # ==========================================
        # Heavy multitasking
        # ==========================================

        if user.multitasking >= 8:
            weights["ram"] += 0.10
            weights["multitask"] += 0.10

        # ==========================================
        # Video editing
        # ==========================================

        if user.video_editing >= 7:
            weights["gpu"] += 0.10
            weights["thermal"] += 0.10

        # ==========================================
        # Portability focus
        # ==========================================

        if user.portability >= 8:
            weights["portability"] += 0.15

        # ==========================================
        # Long lifespan target
        # ==========================================

        if user.lifespan_years >= 6:
            weights["ram"] += 0.10

        return weights


# =========================================================
# SCORING ENGINE
# =========================================================

class ScoringEngine:

    @staticmethod
    def calculate_score(mac: MacBook, user: UserProfile):

        weights = LogicEngine.determine_priorities(user)

        score = 0

        # =====================================================
        # Core weighted scores
        # =====================================================

        score += mac.cpu_score * weights["cpu"]
        score += mac.gpu_score * weights["gpu"]

        score += mac.portability_score * weights["portability"]
        score += mac.battery_score * weights["battery"]

        score += mac.ai_score * weights["ai"]
        score += mac.multitask_score * weights["multitask"]
        score += mac.thermal_score * weights["thermal"]

        # =====================================================
        # RAM logic
        # =====================================================

        ram_score = 0

        if mac.ram >= 24:
            ram_score = 100
        elif mac.ram >= 16:
            ram_score = 75
        else:
            ram_score = 40

        score += ram_score * weights["ram"]

        # =====================================================
        # Budget penalty
        # =====================================================

        if mac.price > user.budget:
            over_budget = mac.price - user.budget
            penalty = over_budget / 100
            score -= penalty

        # =====================================================
        # Value-per-dollar bonus
        # =====================================================

        value_ratio = score / mac.price
        score += value_ratio * 1000

        return round(score, 2)


# =========================================================
# RECOMMENDATION SYSTEM
# =========================================================

class RecommendationEngine:

    @staticmethod
    def recommend(user: UserProfile):

        results = []

        for mac in MACBOOKS:
            score = ScoringEngine.calculate_score(mac, user)

            results.append({
                "model": mac.name,
                "price": mac.price,
                "ram": mac.ram,
                "score": score
            })

        results.sort(key=lambda x: x["score"], reverse=True)

        return results


# =========================================================
# DISPLAY FUNCTIONS
# =========================================================

def print_results(results):

    print("\n" + "=" * 60)
    print("MACBOOK RECOMMENDATION RESULTS")
    print("=" * 60)

    for idx, item in enumerate(results, start=1):

        print(f"""
#{idx}
Model : {item['model']}
Price : ¥{item['price']}
RAM   : {item['ram']}GB
Score : {item['score']}
""")


# =========================================================
# MAIN PROGRAM
# =========================================================

def main():

    """
    User profile example:
    """

    user = UserProfile(

        budget=9000,

        coding=9,
        ai=8,
        video_editing=5,
        multitasking=9,
        portability=8,
        gaming=4,

        lifespan_years=6
    )

    results = RecommendationEngine.recommend(user)

    print_results(results)

    print("\nTOP RECOMMENDATION:")
    print("=" * 60)
    print(results[0]["model"])


# =========================================================

if __name__ == "__main__":
    main()
```
